# RotateSH

```python
import numpy as np
import torch as th
import pytorch_lightning as pl
pl.seed_everything(47)

sj_id = list(range(1, 21))
vs_pair = ['difareli++-vs-ic_light', 
           'difareli++-vs-hou21_shadowm', 
           'difareli++-vs-hou22_geom', 
           'difareli++-vs-diffusionrig',
           'difareli++-vs-relipa'
        ]

# Many-to-Many for sj_id and vs_pair
sj_dict = {}
for sj in sj_id:
    sj_dict[f'pair{sj}'] = [vs for vs in vs_pair]
print("[#] Length of sj_dict: ", len(sj_dict))
print("[#] Each sj has vs_pair: ", len(sj_dict['pair1']))
print("Example: ", list(sj_dict.items())[0])
print("Example: ", list(sj_dict.items())[1])

# Create Global Chunks that each chunk has n sj and keep iterate until that sj has empty vs, then remove that sj
n_sj_total = len(sj_dict)
n_sj_per_chunk = 10
out = []
total_elem = sum([len(v) for v in sj_dict.values()])
print("[#] Total elements: ", total_elem)

all_chunks = []
n_round = total_elem // n_sj_per_chunk // (n_sj_total // n_sj_per_chunk)
for r in range(n_round):
    # Each round we create a chunk
    # Since we have 20 sj, we iterate 2 times to fill 10 sj
    keys = list(sj_dict.keys())
    np.random.shuffle(keys)
    keys_iters = [keys[:n_sj_per_chunk], keys[n_sj_per_chunk:]]

    for i in range(n_sj_total // n_sj_per_chunk):
        chunk = []
        print(f"[#] Round {r} - {i}: ", keys_iters[i])
        for k in keys_iters[i]:
            v = np.random.choice(sj_dict[k], size=1, replace=False)[0]
            chunk.append( (k, v) )
            sj_dict[k].remove(v)
            if len(sj_dict[k]) == 0:
                sj_dict.pop(k, None)

            
        all_chunks.append(chunk)
print("[#] Total chunks: ", len(all_chunks))
print("Total elements in all chunks: ", sum([len(c) for c in all_chunks]))
print("Example chunk: ", all_chunks[0])
print("Example chunk: ", all_chunks[1])
print("Example chunk: ", all_chunks[-1])
print("Final sj_dict: ", sj_dict)

# Write each round to csv file
import pandas as pd
out = {}
out['sj_name'] = []
out['vs_name'] = []
out['radioGroupName'] = []
for r, chunk in enumerate(all_chunks):
    pp = [c[0] for c in chunk]
    vs = [c[1] for c in chunk]
    print(r, chunk)
    pp = '#'.join(pp)
    vs = '#'.join(vs)
    out['sj_name'].append(pp)
    out['vs_name'].append(vs)
    
assert len(out['sj_name']) == len(out['vs_name'])
out['scenario'] = [1] * len(out['sj_name'])
out['question'] = [1] * len(out['sj_name'])
out['radioGroupName'] = [1] * len(out['sj_name'])
pd.DataFrame(out).to_csv('./rotateSH.csv', index=False, columns=['scenario', 'question', 'sj_name', 'vs_name', 'radioGroupName'])
```
